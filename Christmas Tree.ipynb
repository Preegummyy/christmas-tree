```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(10, 10), dpi=300)
ax = fig.add_subplot(111, projection="3d")

# leaf
k = 300
leaf_z = [i for i in range(k)]
leaf_x = [np.cos(i / 5) * (k - i) for i in range(k)]
leaf_y = [np.sin(i / 5) * (k - i) for i in range(k)]

Z = leaf_z
X = leaf_x
Y = leaf_y
ax.scatter(X, Y, Z, c="green", marker="^")

# fill leaf
step = 50
for i in range(k):
    if i==0:
        fill_leaf_size = 500
        U1 = np.random.uniform(size=fill_leaf_size)
        U2 = np.random.uniform(size=fill_leaf_size)
        Z = 0
        X = k * np.sqrt(U2) * np.cos(2 * np.pi * U1)
        Y = k * np.sqrt(U2) * np.sin(2 * np.pi * U1)
        ax.scatter(X, Y, Z, c="green", marker="^")
    if i % step == 0:
        fill_leaf_size = 1 * (i)
        U1 = np.random.uniform(size=fill_leaf_size)
        U2 = np.random.uniform(size=fill_leaf_size)
        Z = k - i
        X = i * np.sqrt(U2) * np.cos(2 * np.pi * U1)
        Y = i * np.sqrt(U2) * np.sin(2 * np.pi * U1)
        ax.scatter(X, Y, Z, c="green", marker="^")

# trunk
xy = np.linspace(0, 2 * np.pi, 20)
zz = np.linspace(-75, 0, 50)
xy, Z = np.meshgrid(xy, zz)
X = 50 * np.cos(xy)
Y = 50 * np.sin(xy)
ax.scatter(X, Y, Z, color="saddlebrown", marker="d")

# star
Z = leaf_z[-1] + 10
X = leaf_x[-1]
Y = leaf_y[-1]
ax.scatter(X, Y, Z, c="yellow", edgecolors="gold", s=200, marker=(5, 1))

# red ball
select = np.random.choice(a=[True, False], size=(300), p=[0.1, 0.9])
red_ball_z = (np.array(leaf_z) - 10)[select]
red_ball_x = np.array(leaf_x)[select]
red_ball_y = np.array(leaf_y)[select]
Z = red_ball_z
X = red_ball_x
Y = red_ball_y
ax.scatter(X, Y, Z, c="red", s=50, marker="o", zorder=2)

# yellow ball
select = np.random.choice(a=[True, False], size=(300), p=[0.1, 0.9])
yellow_ball_z = (np.array(leaf_z) - 10)[select]
yellow_ball_x = np.array(leaf_x)[select]
yellow_ball_y = np.array(leaf_y)[select]
Z = yellow_ball_z
X = yellow_ball_x
Y = yellow_ball_y
ax.scatter(X, Y, Z, c="gold", s=50, marker="o", zorder=2)
ax.view_init(elev=10.5, azim=45)

# snow
Z = np.random.random(size=100) * (300 - (-50)) - 50
X = np.random.random(size=100) * (400 - (-400)) - 400
Y = np.random.random(size=100) * (400 - (-400)) - 400
ax.scatter(X, Y, Z, c="lightskyblue", marker="h")

plt.xlim(-500, 500)
plt.ylim(-500, 500)
plt.show()
```
